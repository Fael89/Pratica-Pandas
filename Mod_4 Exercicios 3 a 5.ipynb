## 2) Vestibular II

Ainda sobre o mesmo banco de dados:

1. Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo *DataFrame* com os alunos desqualificados por este critério.
2. Quantos são esses alunos?
3. Qual a média desses alunos em história e geografia?
4. Monte um *DataFrame* com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.

````python
# 1 - Neste vestibular, quem 'zera' em matemática, física ou química está desqualificado. Monte um novo DataFrame com os alunos desqualificados por este critério. 
df_reprovados = acertos[(df_mat['Qt_acertos'] == 0)|
                       (df_fis['Qt_acertos'] == 0) |
                       (df_qui['Qt_acertos'] == 0)]

df_reprovados
````

````python
# modelo mais detalhado do exercicio acima
#Matematica

df_mat['matematica'] = df_mat['Qt_acertos']
df_mat.drop(['Qt_acertos'], axis='columns', inplace=True)

#Portugues
df_por['portugues'] = df_por['Qt_acertos']
df_por.drop(['Qt_acertos'], axis='columns', inplace=True)

#Geografia
df_geo['geografia'] = df_geo['Qt_acertos']
df_geo.drop(['Qt_acertos'], axis='columns', inplace=True)

#Ingles
df_ing['ingles'] = df_ing['Qt_acertos']
df_ing.drop(['Qt_acertos'], axis='columns', inplace=True)

#Historia
df_his['historia'] = df_his['Qt_acertos']
df_his.drop(['Qt_acertos'], axis='columns', inplace=True)

#Fisica
df_fis['fisica'] = df_fis['Qt_acertos']
df_fis.drop(['Qt_acertos'], axis='columns', inplace=True)

#quimica
df_qui['quimica'] = df_qui['Qt_acertos']
df_qui.drop(['Qt_acertos'], axis='columns', inplace=True)

````

````python
# construindo novo dataframe mais detalhado
df_concat = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1, join='outer')
df_concat
````

````python
# separando quem zerou em matematica, fisica e quimica

desc_mat = df_concat.loc[(df_concat['matematica'] == 0)]
desc_fis = df_concat.loc[(df_concat['fisica'] == 0)]
desc_qui = df_concat.loc[(df_concat['quimica'] == 0)]
````

````python
# criando df dos desclassificados

df_desclassificados = pd.concat([desc_mat, desc_fis, desc_qui], axis=0, join='outer')
df_desclassificados.drop_duplicates(inplace=True)

df_desclassificados
````

````python
# 2 - Quantos são esses alunos?
df_desclassificados.shape[0]
````

````python
# 3 - Qual a média desses alunos em história e geografia?

df_hist_media = df_desclassificados['historia'].mean()

df_hist_media
````

````python
df_desclassificados['geografia'].mean()
````

````python
# 4 - Monte um DataFrame com os alunos que passaram para a segunda fase. Repare que estes alunos não podem ter sido desqualificados.
# Unindo Df Aprovados e Df eliminados
df_aprovados = df_concat[(df_concat['matematica'] != 0) &  (df_concat['fisica'] != 0) & (df_concat['quimica'] != 0)]
df_aprovados
                         
````

````python
# 4 - a partir da lista de colunas, escolha somente aquelas que são referentes ao estabelecimento, faça uma lista com esses valores.
lista_variaveis = menor_df.loc[:,'estabelecimento_valor':'estabelecimento_uf'].columns.unique()
lista_variaveis
````

````python
# 5 - usando o método .loc, selecione somente essas variáveis
lista_var = ['estabelecimento_valor',
       'estabelecimento_razaosocial', 'estalecimento_nofantasia',
       'estabelecimento_municipio_codigo', 'estabelecimento_municipio_nome',
       'estabelecimento_uf']
df_lista_menor = menor_df.loc[:,lista_var]
df_lista_menor
````
