```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_sinasc = pd.read_csv('sinasc.csv')
df_sinasc.columns
```

```python
df_sinasc2 = df_sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']]
df_sinasc2.head(5)
```

```python
df_sinasc2['APGAR1'].value_counts()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# GRAFICO APGAR1
plt.rc('figure', figsize=(10,8))
figure, axes = plt.subplots()

sns.countplot(ax = axes, x='APGAR1', data=df_sinasc2);
```

```python
# APGAR5
plt.rc('figure', figsize=(10,8))
figure, axes = plt.subplots()

sns.countplot(ax = axes, x='APGAR5', data=df_sinasc2);
```

```python
# SEXO
plt.rc('figure', figsize=(8,5))
figure, axes = plt.subplots()

sns.countplot(ax = axes, x='SEXO', data=df_sinasc2);
```

```python
# GESTACAO
plt.rc('figure', figsize=(10,8))
figure, axes = plt.subplots()

sns.countplot(ax = axes, x='GESTACAO', data=df_sinasc2);
```

```python
plt.rc('figure', figsize=(10,8))
figure, axes = plt.subplots()

sns.countplot(ax=axes, x='ESTCIVMAE', data=df_sinasc2)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure', figsize=(10,8))

sns.barplot(x='APGAR5',y='CONSULTAS', data=df_sinasc2);
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(15,8))
fig, axes = plt.subplots(2,2)

sns.countplot(ax = axes[0,0], x='SEXO', data=df_sinasc2)
sns.countplot(ax = axes[0,1], x='GESTACAO', data=df_sinasc2)
sns.countplot(ax = axes[1,0], x='ESTCIVMAE', data=df_sinasc2)
sns.countplot(ax = axes[1,1], x='CONSULTAS', data=df_sinasc2);
```
